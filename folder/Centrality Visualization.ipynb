```python
import folium
import openrouteservice as ors

# create map centered on Heidelberg
m = folium.Map(location = [49.411085, 8.685894], zoom_start = 13)

# request centrality from ors
host = "http://localhost:8082/ors"
profile = "driving-car"
neuenheim = [[8.677139,49.412872],[8.690443,49.421080]]
neuenheim_all = [[neuenheim[0][1], neuenheim[1][0]], [neuenheim[1][1], neuenheim[0][0]], [neuenheim[0][1], neuenheim[0][0]], [neuenheim[1][1], neuenheim[1][0]]]

coords = [[8.676581, 49.418204], [8.692803, 49.409465]]
client = ors.Client(base_url=host, key='5b3ce3597851110001cf6248554c879acee241f6bc733402d2d351be')
resp = client.request(url = f'/v2/centrality/{profile}/json', get_params = {}, post_json = {'bbox': neuenheim})

# visualize rectangle
folium.Rectangle(bounds=neuenheim_all, color='#fcba03', fill=True, fill_color='#fcba03').add_to(m)

# display centrality on map
for loc in resp['locations']:
    loc[0], loc[1] = loc[1], loc[0]
    folium.CircleMarker(location = loc, radius = 1, color='#3186cc').add_to(m)
    
m 
```
